##### Estrutura de um programa Python

 - Script(*.py) vs. Notebook(*.ipynb): Python pode ser executado como um script ou em modo notebook no Jupyter.
 - Comentários: Usar # para comentários de linha única e """ """ ou ''' ''' para comentários de múltiplas linhas.
 - Indentação: Essencial para definir blocos de código, como em funções, loops e condicionais.
 - Variáveis e Tipos de Dados

 - Variáveis: Armazenamento de valores usando nomes descritivos. (cuidado com palavras reservadas)
 - Tipos de dados básicos: Inteiros (int), pontos flutuantes (float), strings (str), booleanos (bool).
 - Conversão de tipos: int(), float(), str(), bool() para converter entre tipos.

 - Operadores Básicos
     - Aritméticos: +, -, *, /, // (divisão inteira), % (módulo), ** (exponenciação).
     - Comparação: ==, !=, >, <, >=, <=.
     - Lógicos: and, or, not.

##### Entrada e Saída de Dados

 - Entrada de dados: input() para capturar dados do usuário.
 - Saída de dados: print() para exibir dados na tela.


##### Condicionais (if, elif, else)

Estruturas de decisão para executar diferentes blocos de código com base em condições.

 - if: Executa um bloco de código se a condição for verdadeira.
 - elif: Executa um bloco de código se a condição anterior for falsa e a condição atual for verdadeira.
 - else: Executa um bloco de código se todas as condições anteriores forem falsas.


##### Laços de Repetição (for, while)

 - For: Iteração sobre sequências (listas, tuplas, strings, etc.).
 - While: Repetição baseada em uma condição booleana.
 - Controle de Laços (break, continue)
     - Break: Interrompe a execução do loop.
     - Continue: Pula para a próxima iteração do loop.


##### Funções em Python
    
- Definição e Chamada de Funções
    - Definir funções com def nome_da_funcao():.
    - Chamar funções para executar o código definido.
- Parâmetros e Retorno de Funções
    - Passagem de argumentos para funções.
    - Uso de return para devolver valores de funções.
- Escopo de Variáveis
    - Escopo local vs. escopo global.
    - Uso de global para modificar variáveis globais dentro de funções.


##### Estruturas de Dados Básicas

Strings e Suas Operações

 - Operações comuns: concatenação, fatiamento, métodos (upper(), lower(), split(), etc.).

 - Listas: Estruturas mutáveis, métodos comuns (append(), remove(), pop(), etc.).

 - Tuplas: Estruturas imutáveis, usos e vantagens.

 - Dicionários: Estruturas chave-valor, métodos (keys(), values(), items(), etc.).

 - Conjuntos: Estruturas não ordenadas de elementos únicos, operações (add(), remove(), união, interseção, etc.).


##### Módulos e Bibliotecas

Importação e Utilização de Módulos

Uso de import para incluir módulos padrão e personalizados.
Exploração de Bibliotecas Padrão: Exemplos como math e random.

Uso de pip para instalar pacotes externos.


##### Classes

As classes são a base da programação orientada a objetos em Python.

 - Definição de Classes
    - Definir classes com class NomeDaClasse:.
    - Inicialização de objetos com 'def __init__'.
 - Atributos e Métodos
    - Atributos: Variáveis associadas a objetos.
    - Métodos: Funções associadas a objetos.
 - Herança
    - Criar classes filhas que herdam atributos e métodos de classes pai.
 - Encapsulamento
    - Proteger atributos e métodos de acesso externo.
 - Polimorfismo
    - Utilizar métodos com o mesmo nome em classes diferentes.


##### Tratamento de Exceções, Boas Práticas e Documentação

 - Try, Except, Else, Finally: Estruturas para capturar e tratar erros.
 - Raise: Lançar exceções personalizadas.
 - Assert: Verificações de condições esperadas.


##### Boas Práticas PEP8

Convenções de estilo de código para manter a legibilidade e consistência.
Documentação

##### PEP257 para convenções de docstrings.

Convenções de documentação para funções, classes e módulos.


##### Manipulação de Arquivos e Sistema
Leitura e Escrita de Arquivos

Funções open(), read(), write(), close() para manipulação de arquivos.


##### Manipulação de Diretórios

Uso do módulo os para navegação e manipulação de diretórios e arquivos.

 - os.chdir(), os.listdir(), os.mkdir(), os.rmdir(), etc.
 - os.path para manipulação de caminhos de arquivos.
